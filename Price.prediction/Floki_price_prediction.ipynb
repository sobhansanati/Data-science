```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from seaborn import regression
```

```python
sns.set()
plt.style.use('seaborn-whitegrid')

data = pd.read_csv('FLOKI-USD.csv')
data.head()
```

```python
data.dropna()
plt.figure(figsize=(18,8))
plt.title('Floki inu price')
plt.xlabel('Date')
plt.ylabel('Close')
plt.plot(data["Close"])
plt.show()
```

```python
from autots import AutoTS # a automated meachine learning system for time series analysis T D V

model = AutoTS(forecast_length=30,frequency='infer',ensemble='simple',drop_data_older_than_periods=200)
model = model.fit(data,date_col='Date',value_col="Close",id_col=None)
```

```python
predictions = model.predict()
forecast = predictions.forecast
print('Floki inu Price prediction ')
print(forecast)
```
