```python
import numpy as np
import pandas as pd
import string
import re
import tensorflow as tf
import matplotlib.pyplot as plt
```

# Prepare Data

```python
from google.colab import drive
drive.mount('D')
```

```python
train_path = 'train.csv'
test_path = 'test.csv'
data = pd.read_csv(train_path)
col = ['id', 'comment', 'price_value', 'fake_originality', 'warranty', 'size', 'discrepancy', 'flavor_odor', 'expiration_date']
data = data[col]
print(data.shape)
data.head()
```

```python
def clean_txt(text):

    text = text.replace('\u200c',' ')
    text = re.sub('https:\/\/\S+', ' ', text) 
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) 
    text = re.sub(r'[^ \w\.]', ' ', text) 
    text = re.sub('\w*\d\w*', ' ', text)
    text = re.sub(r'[^a-zA-Z0-9آ-ی۰-۹ ]', ' ', text)
    return text
```

```python
data['comment'] = data['comment'].astype('str')
data['comment'] = data['comment'].apply(lambda x: clean_txt(x))
data.head()
```

```python
class tokenizer():
    def __init__(self, data):
        
        #data: A list ar Array from texts
        self.data = data
        self.set_of_words = set()
        for txt in data:
            [self.set_of_words.add(x) for x in txt.split()]
            
    def fit(self):
        self.word_to_id = dict()
        for idx, word in enumerate(self.set_of_words):
            self.word_to_id.update({word:idx})

    def transform(self, inp):
        
        transformed_txt = list()
        for txt in inp:
            temp = list()
            for word in txt.split():
              try:
                temp.append(self.word_to_id[word])
              except:
                continue;
            transformed_txt.append(temp)
        
        return transformed_txt
```

```python
T = tokenizer(data.comment.values)
T.fit()
data.comment = T.transform(data.comment.values)
data.head()
```

```python
truncated_txt = list()

for i in range(data.shape[0]):
    cmnt = np.zeros(256, dtype=np.float32)
    try:
        cmnt[:len(data.iloc[i]['comment'])] = data.iloc[i]['comment']
    except:
        cmnt = data.iloc[i]['comment'][:256]
    
    cmnt = tf.convert_to_tensor(cmnt, dtype='float32')
    truncated_txt.append(cmnt)
```

```python
truncated_txt = tf.convert_to_tensor(truncated_txt, dtype=tf.float32)

price_value = tf.convert_to_tensor(data.price_value.values, dtype=tf.int32)
price_value = tf.one_hot(price_value, 2, dtype=tf.float32)

fake_originality = tf.convert_to_tensor(data.fake_originality.values, dtype=tf.int32)
fake_originality = tf.one_hot(fake_originality, 2, dtype=tf.float32)

warranty = tf.convert_to_tensor(data.warranty.values, dtype=tf.int32)
warranty = tf.one_hot(warranty, 2, dtype=tf.float32)

size = tf.convert_to_tensor(data['size'].values, dtype=tf.int32)
size = tf.one_hot(size, 2, dtype=tf.float32)

discrepancy = tf.convert_to_tensor(data.discrepancy.values, dtype=tf.int32)
discrepancy = tf.one_hot(discrepancy, 2, dtype=tf.float32)

flavor_odor = tf.convert_to_tensor(data.flavor_odor.values, dtype=tf.int32)
flavor_odor = tf.one_hot(flavor_odor, 2, dtype=tf.float32)

expiration_date = tf.convert_to_tensor(data.expiration_date.values, dtype=tf.int32)
expiration_date = tf.one_hot(expiration_date, 2, dtype=tf.float32)
```

# Define Model

```python
class joint_model(tf.keras.Model):

    def __init__(self):
        super().__init__()
        
        self.embed = tf.keras.layers.Embedding(len(T.set_of_words) + 1,  128)
        self.lstm_layer = tf.keras.layers.LSTM(512, return_sequences=True)
        
        self.out_1 = tf.keras.layers.Dense(2, "softmax")
        self.out_2 = tf.keras.layers.Dense(2, "softmax")
        self.out_3 = tf.keras.layers.Dense(2, "softmax")
        self.out_4 = tf.keras.layers.Dense(2, "softmax")
        self.out_5 = tf.keras.layers.Dense(2, "softmax")
        self.out_6 = tf.keras.layers.Dense(2, "softmax")
        self.out_7 = tf.keras.layers.Dense(2, "softmax")

    def call(self, x):
        
        x = self.embed(x)
        x = self.lstm_layer(x)
        x = tf.keras.layers.Activation(tf.nn.relu)(x)
        x = tf.keras.layers.Dropout(0.4)(x)
        x = tf.keras.layers.Flatten()(x)
        
        out_1 = self.out_1(x)
        out_2 = self.out_2(x)
        out_3 = self.out_3(x)
        out_4 = self.out_4(x)
        out_5 = self.out_5(x)
        out_6 = self.out_6(x)
        out_7 = self.out_7(x)

        return out_1, out_2, out_3, out_4, out_5, out_6, out_7
```

```python
model = joint_model()

opt = tf.keras.optimizers.Adam(learning_rate=3e-4, epsilon=1e-08)

losses = [tf.keras.losses.CategoricalCrossentropy(),
          tf.keras.losses.CategoricalCrossentropy(),
          tf.keras.losses.CategoricalCrossentropy(),
          tf.keras.losses.CategoricalCrossentropy(),
          tf.keras.losses.CategoricalCrossentropy(),
          tf.keras.losses.CategoricalCrossentropy(),
          tf.keras.losses.CategoricalCrossentropy()]

metrics = ["accuracy"]

# compile model
model.compile(optimizer=opt, loss=losses, metrics=metrics)
```

```python
model.fit(truncated_txt, (price_value, fake_originality, warranty, size, discrepancy, flavor_odor, expiration_date),
          validation_split=0.2,
          batch_size=512, epochs=50)
```

## Prepare Test Data

```python
data = pd.read_csv(test_path)
col = ['id', 'comment']
data = data[col]
print(data.shape)
data.head()
```

```python
data['comment'] = data['comment'].astype('str')
data['comment'] = data['comment'].apply(lambda x: clean_txt(x))
data.head()
```

```python
data.comment = T.transform(data.comment.values)
data.head()
```

```python
truncated_txt = list()

for i in range(data.shape[0]):
    cmnt = np.zeros(256, dtype=np.float32)
    try:
        cmnt[:len(data.iloc[i]['comment'])] = data.iloc[i]['comment']
    except:
        cmnt = data.iloc[i]['comment'][:256]
    
    cmnt = tf.convert_to_tensor(cmnt, dtype='float32')
    truncated_txt.append(cmnt)

truncated_txt = tf.convert_to_tensor(truncated_txt, dtype=tf.float32)
```

### Predict labels

```python
predictions = list()
for i in range(data.shape[0]):
  temp = list()
  temp.append(data.iloc[i]['id'])
  preds = model.predict(truncated_txt[i][np.newaxis, ...], verbose=False)
  for item in preds:
    temp.extend(np.argmax(item, axis=-1))
  predictions.append(temp)
```

```python
preds = pd.DataFrame(data=predictions, columns = ['id', 'price_value', 'fake_originality', 'warranty', 'size', 'discrepancy', 'flavor_odor', 'expiration_date'])
print(preds.shape)
preds.head()
```
