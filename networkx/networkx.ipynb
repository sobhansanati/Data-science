```python
import networkx as nx 
import matplotlib as plt
```

```python
G = nx.complete_graph(9)
nx.draw_networkx(G)
```

```python
g=nx.Graph()
g.add_node('A')
g.add_node('B')
g.add_nodes_from([
    (4, {"color": "red"}),
    (5, {"color": "green"}),
    ])
g.add_edges_from([('A', 'B'), (4, 5)])
print(g.number_of_nodes())
print(g.number_of_edges())
print(g.nodes())
print(g.edges())
nx.draw_networkx(g)
```

```python
G = nx.Graph()
 
# Adding one edge at a time
# Node 1 and 2 will be automatically added
G.add_edge(1,2)
G.add_edge(3,2)
 
# Adding multiple edges at a time
G.add_edges_from([(4,2), (3,5), (5,4)])
 
# Adding duplicates will be ignored.
G.add_node(1)
G.add_edge(1,2)
 
print(G.nodes())
print(G.edges())

for i in range(1,6):
    print(G.degree(i))
G.remove_node(1)    
nx.draw_networkx(G)
```

```python
G = nx.Graph()
 
# Using add_edge
G.add_edge(1, 2, weight = 12.5)
G.add_edge(3, 2, weight = 50.0)
G.add_edge(1, 3, weight = 17)
G.add_edge(4, 2, weight = 100)
G.add_edge(2, 5, weight = 1)
G.add_edge(4, 6, weight = 25.5)
G.add_edge(7, 4, weight = 175)
G.add_edge(5, 8, weight = 90)
 
nx.draw(G, with_labels= True, font_weight='bold')
```

```python
G = nx.Graph()
 
# Using add_edge
G.add_edge(1, 2, weight = 12.5)
G.add_edge(3, 2, weight = 50.0)
G.add_edge(1, 3, weight = 17)
G.add_edge(4, 2, weight = 100)
G.add_edge(2, 5, weight = 1)
G.add_edge(4, 6, weight = 25.5)
G.add_edge(7, 4, weight = 175)
G.add_edge(5, 8, weight = 90)
 
 
pos=nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold')
edge_weight = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels = edge_weight)
plt.show()
for i in range(1,9):
    print(nx.clustering(G,i))
print(nx.average_clustering(G))
```
