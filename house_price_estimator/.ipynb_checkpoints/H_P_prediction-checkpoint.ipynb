```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_train=pd.read_csv("/content/train.csv")
data_train
```

```python
data_test=pd.read_csv("/content/test.csv")
data_test_copy=data_test.copy()
data_test
```

```python
nan_col=data_train.columns[data_train.isna().sum()/len(data_train)>0.25]
nan_col
```

```python
data_train.drop(columns=nan_col,inplace=True)
data_test.drop(columns=nan_col,inplace=True)
data_train
```

```python
data_train.drop(columns=["Id"],inplace=True)
data_test.drop(columns=["Id"],inplace=True)

data_train
```

```python
pd.set_option('max_columns', None)
data_train
```

```python
categorical_columns = (data_train.dtypes == 'object')
categorical_columns=data_train.columns[categorical_columns]
len(categorical_columns)
```

```python
categorical_data_train=data_train[categorical_columns]
categorical_data_test=data_test[categorical_columns]

```

```python

dummies_data_train=pd.get_dummies(categorical_data_train)

dummies_data_test=pd.get_dummies(categorical_data_test)

dummies_data_test = dummies_data_test.reindex(columns = dummies_data_train.columns, fill_value=0)
dummies_data_test
```

```python
data_train.drop(columns=categorical_columns, inplace=True)
data_test.drop(columns=categorical_columns, inplace=True)

data_train
```

```python
price=data_train["SalePrice"].values
data_train.drop(columns=["SalePrice"],inplace=True)

```

```python
data_train.fillna(data_train.mean(),inplace=True)#filling with mean
data_test.fillna(data_test.mean(),inplace=True)#filling with mean
```

```python
data_train=pd.concat([data_train,dummies_data_train],axis=1)
data_test=pd.concat([data_test,dummies_data_test],axis=1)

data_train
```

```python
data_test
```

```python
data_train.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(data_train)
data_train=scaler.transform(data_train)
data_test=scaler.transform(data_test)
```

```python
max_price=max(price)
price=price/max_price
```

```python
from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val=train_test_split(data_train,price,test_size=0.2,random_state=2,shuffle=True)
```

```python
from keras.models import Sequential # Sequential Neural Network
from keras.layers import Dense
```

```python
x_train.shape
```

```python

model = Sequential()
model.add(Dense(64, input_dim=x_train.shape[1], activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))

model.add(Dense(1))
    # Compile model
model.compile(optimizer="adadelta", loss = 'mse')
 
model.summary()
```

```python
history = model.fit(
    x_train, y_train,
    validation_data = (x_val, y_val),
    epochs = 100,

)
```

```python
train_loss=history.history["loss"]
val_loss=history.history["val_loss"]
EPOCHS=range(len(train_loss))
plt.plot(EPOCHS,train_loss,
        EPOCHS, val_loss)
```

```python
test_predict=model.predict(data_test)
test_predict_values=test_predict*max_price
```

```python
test_predict_values
```

```python
data_test_copy["SalePrice"]=test_predict_values
data_test_copy
```

```python
data_test_copy.to_csv("test_with_price.csv")
```
